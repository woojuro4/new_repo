```python
import konlpy
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
txt = open('text.txt', encoding='utf8').read()
txt
```

```python
tokens_noun = okt.nouns(txt)
tokens_noun
```

```python
from collections import Counter
nouns_dict=dict(Counter(tokens_noun).most_common(1000))
nouns_dict
```

```python
wordcloud = WordCloud(background_color = "white",
                      font_path = 'C:/Windows/Fonts/malgun.ttf')

wordcloud.generate_from_frequencies(nouns_dict)
wordcloud.to_image()
```
